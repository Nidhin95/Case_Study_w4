# 3. Bar graph of male vs female buyers participated in the sales.

```python
#grouping male data
male=data[data['Buyer Gender'] == 'Male']
```

```python
#male
```

```python
#grouping female data
female=data[data['Buyer Gender'] == 'Female']
```

```python
#female
```

```python
plt.figure(figsize=(8,8))
plt.hist(male['Buyer Gender'],rwidth=1,color='green')
plt.hist(female['Buyer Gender'],rwidth=1)
plt.ylabel('Number Of Buyers',fontsize=15)
plt.xlabel('Buyer Gender',fontsize=15)
plt.title('Male vs Female Buyer Participation',fontsize=20)
#adding legend manually
plt.legend(('male','female'),loc=[1.02,.9])
plt.grid(color='red',linewidth=1.05,axis='y',alpha=2)
```

```python
Insights:
    
    1.Total number of female buyers > Total number of male buyers.
    2.Average numer of buyers is around 5000.
```

```python
plt.figure(figsize=(9,9))
plt.title('Age Distribution Among Buyers(Male VS Female)\n',fontsize=18)
plt.hist(data['Buyer Age'],rwidth=.5,alpha=.5)
plt.hist(male['Buyer Age'],rwidth=.5)
plt.hist(female['Buyer Age'],rwidth=.3)

plt.xlabel('\nAGE',fontsize=15)
plt.ylabel('Number of Buyers',fontsize=15)
plt.grid(color='red',linewidth=1.5,axis='y',alpha=2)
plt.legend(('Total','Male','Female'))
```

```python
Insights:
    
    1. More buyers are from above 70 category
    2. The least number of buyers are in the age group of 25-30 years. 
    3. In 5 cases the number of female buyers is greater than the male buyers
    4. only in 3 cases the number of buyers is < 900
```

# 4. Finding the top 5 cars based on their sales price.

```python
data.Model
```

```python
#identifying cars with highest Sale price and grouping
sale=data[data['Sale Price'] > 99950]
sale
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x='Model',y='Sale Price',s=10,data=sale)
sns.lineplot(x='Model',y='Sale Price',data=sale)
plt.xlabel('\nCar Models',fontsize=15)
plt.title('Top Cars VS Sales Price\n',fontsize=18)
plt.grid(color='black',linewidth=1.5,axis='y',alpha=.2)
plt.legend(('Sale Price','5th','2nd','3rd','7th','1st','6th','4th'),loc=[1.02,0])
```

```python
Insights:
    
    1.There are total 7 car models having Sale price > 99950.
    2.Top 5 Car Models are Jetta III, A6, LX, MX-6 and ES
    3.Jetta III is having the highest Sale Price among all.
  
NB: insights are made without considering the 'discount'

```

```python
#CONSIDERING DISCOUNT
dat=data
dat['SalePrice']=dat['Sale Price'] - dat['Discount']
```

```python
#dat
```

```python
sale2=dat[dat['SalePrice'] > 99950]
sale2
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x='Model',y='SalePrice',s=10,data=sale2)
sns.lineplot(x='Model',y='SalePrice',data=sale2)
plt.xlabel('\nCar Models',fontsize=15)
plt.title('Top Cars VS Sales Price\n',fontsize=18)
plt.grid(color='black',linewidth=1.5,axis='y',alpha=.2)
```

```python
Insights:
    
    1.There is no chenge in the order while considering the discount as well.
```

# 5.Finding the least 5 cars based on their Resell price

```python
##identifying cars with least resell price and grouping them
re=data[data['Resell Price']<=2040]
```

```python
re
```

```python
plt.figure(figsize=(15,8))
sns.swarmplot(x='Model',y='Resell Price',s=10,data=re)
sns.lineplot(x='Model',y='Resell Price',data=re)
plt.grid(color='black',linewidth=1.5,axis='y',alpha=.2)
plt.xlabel('\nCar Models\n',fontsize=15)
plt.title('Cars With Lowest Resell Price',fontsize=20)
```

```python
Insights:
    
    1. There are 10 car models having resell price < 2040
    2. The list of 5 car models having least RP is 1500, Savana 1500, XF, 3000GT, and G37.
    3. Model 1500 made by Chevrolet is the car with the least Resell price.
    
```
